```python
import urllib.request
from bs4 import BeautifulSoup

URL = urllib.request.urlopen("https://en.wikipedia.org/wiki/Deep_learning")
s = BeautifulSoup(URL, "html.parser")

title = s.find('title')
links = s.find_all('a')

print("Title:", title.text)
print("Here is the links for this page:")
for l in links:
      print(l.get('href'))
```
