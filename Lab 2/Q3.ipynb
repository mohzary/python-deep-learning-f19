# Using CNN model for Image Classification

Dataset: Heart Disease | [download]( https://www.kaggle.com/slothkong/10-monkey-species)

```python
from keras.preprocessing.image import load_img
import matplotlib.pyplot as pltLab2
# Illustrating an example of the traning data
LABELLAB2, IMAGELAB2 = 'n0', 'n0021.jpg'
imgLAB2 = load_img("./10-monkey-species/training/training/n0/n0021.jpg") # image loaded as PIL object
displayLabel = pltLab2.imshow(imgLAB2)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Getting the training and validation folders
TrainingDIRLab2 = './10-monkey-species/training/training'
ValidationDIRLab2 = './10-monkey-species/validation/validation'
RangeLab2 = 0.2
BatchSizeLab2 = 64
imageSizeLab2 = 300

# The generator of the training
TrainDatagenLab2 = ImageDataGenerator(
                                    rescale=1./255,
                                    horizontal_flip = True,
                                    width_shift_range = RangeLab2,
                                    height_shift_range = RangeLab2,
                                    fill_mode='nearest',
                                    shear_range=RangeLab2,
                                    zoom_range=RangeLab2,
                                    rotation_range=40)
# The generator of the testing
TestDatagenLab2 = ImageDataGenerator(rescale=1./255)

TrainGeneratorLab2 = TrainDatagenLab2.flow_from_directory(
                                    directory=TrainingDIRLab2,
                                    shuffle=True,
                                    target_size=(imageSizeLab2, imageSizeLab2),
                                    batch_size=BatchSizeLab2,
                                    seed=1337,
                                    class_mode='categorical')

ValidationGeneratorLab2 = TestDatagenLab2.flow_from_directory(
                                    directory=ValidationDIRLab2,
                                    shuffle=False,
                                    target_size=(imageSizeLab2, imageSizeLab2),
                                    batch_size=BatchSizeLab2,
                                    seed=1337,
                                    class_mode='categorical')
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Dropout, MaxPooling2D
numClassesLab2 = 10

# Creating the model
imageModelLab2 = Sequential()
imageModelLab2.add(Conv2D(32, (3, 3), activation='relu',  input_shape=(300, 300, 3)))
imageModelLab2.add(MaxPooling2D(pool_size=(2, 2)))
imageModelLab2.add(Conv2D(32, (3, 3), activation='relu'))
imageModelLab2.add(MaxPooling2D(pool_size=(2, 2)))
imageModelLab2.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
imageModelLab2.add(Conv2D(64, (3, 3), activation='relu'))
imageModelLab2.add(MaxPooling2D(pool_size=(2, 2)))
imageModelLab2.add(Dropout(0.25))
imageModelLab2.add(Flatten())
imageModelLab2.add(Dense(512, activation='relu'))
imageModelLab2.add(Dropout(0.5))
imageModelLab2.add(Dense(numClassesLab2, activation='softmax'))

imageModelLab2.compile(loss=keras.losses.categorical_crossentropy,
              optimizer='adam',
              metrics=['accuracy'])
```

```python
train_num = TrainGeneratorLab2.samples
validation_num = ValidationGeneratorLab2.samples 
print(train_num)
print(validation_num)
```

```python
# Traning the model
CNNLab2MODEL = imageModelLab2.fit_generator(TrainGeneratorLab2, epochs=100, steps_per_epoch=1097//BatchSizeLab2, validation_data=ValidationGeneratorLab2,
                   validation_steps=272//BatchSizeLab2)
```

```python
imageModelLab2.summary()
```

```python
import matplotlib.pyplot as pltLab2
# plot the acc and val_acc
pltLab2.plot(CNNLab2MODEL.history['acc'])
pltLab2.plot(CNNLab2MODEL.history['val_acc'])
pltLab2.title('model acc')
pltLab2.ylabel('acc')
pltLab2.xlabel('epoch')
pltLab2.legend(['acc', 'val_acc'], loc='upper left')
pltLab2.show()
```

```python
import matplotlib.pyplot as pltLab2
# plot the acc and val_acc
pltLab2.plot(CNNLab2MODEL.history['loss'])
pltLab2.plot(CNNLab2MODEL.history['val_loss'])
pltLab2.title('model acc')
pltLab2.ylabel('loss')
pltLab2.xlabel('epoch')
pltLab2.legend(['loss', 'val_loss'], loc='upper left')
pltLab2.show()
```
